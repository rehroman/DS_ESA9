Excercise 1 Linear Regression

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

Einfaches Beispiel für Lineare Regression:
Umfrage in der Familie nach Körpergröße und Schuhgröße ergibt:

```python
X = [183, 155, 170, 159]
y = [43, 36, 41, 37]
plt.scatter(X, y);
```

```python
slope, intercept, r, p, std_err = stats.linregress(X, y)

def myfunc(X):
 return slope * X + intercept

mymodel = list(map(myfunc, X))

plt.plot(X, mymodel)
plt.scatter(X, y);
plt.ylim(ymin=35.7, ymax=43.5)
plt.xlim(xmin=154, xmax=185)
plt.xlabel("Körpergröße in cm")
plt.ylabel ("Schuhgröße")
plt.show()
```
